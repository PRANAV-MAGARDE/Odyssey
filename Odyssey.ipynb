```python
!pip install astroquery -q

import os
import re
import pandas as pd
from astroquery.simbad import Simbad

# ── Extract catalog name from filename ────────────────────────
def extract_catalog_name(filename):
    match = re.match(r'^(M\d+|NGC\d+|IC\d+)', filename, re.IGNORECASE)
    if match:
        return match.group(1).upper()
    return None

# ── Map SIMBAD type to your label ─────────────────────────────
simbad_map = {
    'G'  : 1,  # Galaxy → spiral default
    'GiG': 1,  # Galaxy in group
    'SyG': 1,  # Seyfert Galaxy
    'H2G': 1,  # HII Galaxy
    'EmG': 1,  # Emission Galaxy
    'AGN': 1,  # Active Galaxy
    'BiC': 0,  # Brightest galaxy in cluster → elliptical
    'Neb': 2,  # Nebula
    'SNR': 2,  # Supernova remnant
    'HII': 2,  # HII region
    'MoC': 2,  # Molecular Cloud
    'RNe': 2,  # Reflection Nebula
    'ISM': 2,  # Interstellar Medium
    'Cl*': 3,  # Star Cluster
    'GlC': 3,  # Globular Cluster
    'OpC': 3,  # Open Cluster
    'As*': 3,  # Association of Stars
    'PN' : 4,  # Planetary Nebula
}

label_names = {0: 'elliptical', 1: 'spiral', 2: 'nebula', 3: 'star_cluster', 4: 'planetary'}

# ── Query SIMBAD ──────────────────────────────────────────────
def get_label(catalog_name):
    try:
        result = Simbad.query_object(catalog_name)
        if result is None:
            # print(f"SIMBAD query for {catalog_name} returned None") # Debug print
            return None
        otype = result['OTYPE'][0]
        # check direct match first
        if otype in simbad_map:
            return simbad_map[otype]
        # check partial match
        for key in simbad_map:
            if key in otype:
                return simbad_map[key]
        # print(f"No matching otype for {catalog_name}: {otype}") # Debug print
        return None
    except Exception as e:
        # print(f"Error querying SIMBAD for {catalog_name}: {e}") # Debug print
        return None

# ── Parse full dataset ─────────────────────────────────────────
def parse_catalog_dataset(images_dir):
    records  = []
    unknown  = []
    files    = os.listdir(images_dir)

    for i, fname in enumerate(files):
        catalog_name = extract_catalog_name(fname)
        if catalog_name is None:
            unknown.append(fname)
            continue

        label = get_label(catalog_name)
        img_path = os.path.join(images_dir, fname)

        if label is not None:
            records.append({'img_path': img_path, 'Label': label, 'catalog': catalog_name})
        else:
            unknown.append(fname)

        # progress every 10 files (SIMBAD queries take time)
        if (i+1) % 10 == 0:
            print(f"Processed {i+1}/{len(files)}...")

    df = pd.DataFrame(records)
    print(f"\nLabeled  : {len(df)}")
    print(f"Unknown  : {len(unknown)}")
    # Only try to print distribution if the DataFrame is not empty
    if not df.empty:
        print(f"\nDistribution:\n{df['Label'].map(label_names).value_counts()}")
    else:
        print("No labeled data to show distribution.")
    return df

# ── Run ───────────────────────────────────────────────────────
test2026_path = '/content/DeepSpaceYoloDatasetV2/DeepSpaceYoloDatasetV2/test2026/images'
deepsky_df    = parse_catalog_dataset(test2026_path)
deepsky_df.to_csv('deepsky_labeled.csv', index=False)
print(deepsky_df.head())
```

```python
import os
import requests
from bs4 import BeautifulSoup
import time

CATEGORIES = {
    "galaxies"     : "https://esahubble.org/images/archive/category/galaxies/",
    "nebulae"      : "https://esahubble.org/images/archive/category/nebulae/",
    "star_clusters": "https://esahubble.org/images/archive/category/star_clusters/",
    "solar_system" : "https://esahubble.org/images/archive/category/solar_system/"
}

BASE_URL           = "https://esahubble.org"
IMAGES_PER_CATEGORY = float('inf')
OUTPUT_BASE        = "/content/hubble_dataset"  # Colab local path

def download_category_images(category_name, category_url):
    output_dir = os.path.join(OUTPUT_BASE, category_name)
    os.makedirs(output_dir, exist_ok=True)

    downloaded_count = 0
    current_page     = 1

    print(f"\n[+] Downloading: {category_name.upper()}")

    while downloaded_count < IMAGES_PER_CATEGORY:
        page_url = f"{category_url}page/{current_page}/" if current_page > 1 else category_url

        response = requests.get(page_url, timeout=100)
        if response.status_code != 200:
            print(f"Status code for {page_url}: {response.status_code} - {response.reason}")
            print(f"No more pages for {category_name}.")
            break

        soup        = BeautifulSoup(response.text, 'html.parser')
        image_links = []

        for img_tag in soup.find_all('img'):
            parent_a = img_tag.find_parent('a')
            if parent_a and '/images/' in parent_a.get('href', ''):
                href = parent_a['href']
                if href not in image_links and not any(x in href for x in ['category', 'viewall', 'search']):
                    image_links.append(href)

        if not image_links:
            print(f"No images found on page {current_page}, stopping.")
            break

        for link in image_links:
            if downloaded_count >= IMAGES_PER_CATEGORY:
                break

            detail_url = BASE_URL + link if link.startswith('/') else link
            try:
                detail_resp = requests.get(detail_url, timeout=10)
                detail_soup = BeautifulSoup(detail_resp.text, 'html.parser')

                # ✅ fixed: search more broadly for image source
                img_src = None
                for selector in [
                    lambda s: s.find('div', class_='main-image-container'),
                    lambda s: s.find('img', class_='main-img'),
                    lambda s: s.find('a', string='Large JPEG'),
                ]:
                    el = selector(detail_soup)
                    if el:
                        img_src = el.get('href') or el.get('src') or (el.find('img').get('src') if el.find('img') else None)
                        if img_src:
                            break

                if not img_src:
                    continue

                img_download_url = BASE_URL + img_src if img_src.startswith('/') else img_src
                filename  = os.path.basename(img_download_url.split('?')[0])
                file_path = os.path.join(output_dir, filename)

                if os.path.exists(file_path):
                    downloaded_count += 1
                    continue

                img_data = requests.get(img_download_url, stream=True, timeout=15)
                with open(file_path, 'wb') as f:
                    for chunk in img_data.iter_content(chunk_size=8192):
                        f.write(chunk)

                downloaded_count += 1
                print(f"  [{downloaded_count}/{IMAGES_PER_CATEGORY}] {filename}")
                time.sleep(0.5)

            except Exception as e:
                print(f"  Error: {e}")
                continue

        current_page += 1

    print(f"  Done: {downloaded_count} images saved to {output_dir}")


# ── Run ───────────────────────────────────────────────────────
for cat_name, cat_url in CATEGORIES.items():
    download_category_images(cat_name, cat_url)

print("\nAll downloads complete!")
print(f"Files saved at: {OUTPUT_BASE}")

# ── Verify ────────────────────────────────────────────────────
for cat in CATEGORIES:
    path  = os.path.join(OUTPUT_BASE, cat)
    count = len(os.listdir(path)) if os.path.exists(path) else 0
    print(f"{cat}: {count} images")
```

```python
import os
import requests
from bs4 import BeautifulSoup
import time

# ✅ correct URLs from inspecting the page
CATEGORIES = {
    "galaxies"    : "https://esahubble.org/images/archive/category/galaxies/",
    "nebulae"     : "https://esahubble.org/images/archive/category/nebulae/",
    "star_clusters": "https://esahubble.org/images/archive/category/starclusters/",
    "solar_system": "https://esahubble.org/images/archive/category/solarsystem/"
}

BASE_URL    = "https://esahubble.org"
OUTPUT_BASE = "/content/hubble_dataset"
IMAGES_PER_CATEGORY = float('inf')

def get_image_links(soup):
    links = []
    for a in soup.find_all('a', href=True):
        href = a['href']
        # individual image pages follow pattern /images/XXXX/
        if href.startswith('/images/') and \
           not any(x in href for x in ['archive', 'category', 'viewall',
                                        'potw', 'potm', 'search', 'zip']):
            if href not in links:
                links.append(href)
    return links

def get_image_src(detail_soup):
    # try to find large jpeg download link
    for a in detail_soup.find_all('a', href=True):
        href = a['href']
        if any(ext in href.lower() for ext in ['.jpg', '.jpeg']):
            if 'large' in href.lower() or 'screen' in href.lower():
                return href

    # fallback: any jpg link
    for a in detail_soup.find_all('a', href=True):
        if any(ext in a['href'].lower() for ext in ['.jpg', '.jpeg']):
            return a['href']

    # fallback: main image tag
    img = detail_soup.find('img', class_='main-img') or \
          detail_soup.find('img', id='main-image')
    if img:
        return img.get('src')

    return None

def download_category_images(category_name, category_url):
    output_dir = os.path.join(OUTPUT_BASE, category_name)
    os.makedirs(output_dir, exist_ok=True)

    downloaded_count = 0
    current_page     = 1

    print(f"\n[+] Downloading: {category_name.upper()}")

    while downloaded_count < IMAGES_PER_CATEGORY:
        page_url = f"{category_url}page/{current_page}/" if current_page > 1 else category_url

        response = requests.get(page_url, timeout=10)
        if response.status_code != 200:
            print(f"No more pages at page {current_page}.")
            break

        soup        = BeautifulSoup(response.text, 'html.parser')
        image_links = get_image_links(soup)

        if not image_links:
            print(f"No image links on page {current_page}, stopping.")
            break

        print(f"  Page {current_page}: found {len(image_links)} images")

        for link in image_links:
            if downloaded_count >= IMAGES_PER_CATEGORY:
                break

            detail_url = BASE_URL + link
            try:
                detail_resp = requests.get(detail_url, timeout=10)
                detail_soup = BeautifulSoup(detail_resp.text, 'html.parser')

                img_src = get_image_src(detail_soup)
                if not img_src:
                    print(f"  No image src found for {link}")
                    continue

                img_download_url = BASE_URL + img_src if img_src.startswith('/') else img_src
                filename  = os.path.basename(img_download_url.split('?')[0])
                file_path = os.path.join(output_dir, filename)

                if os.path.exists(file_path):
                    downloaded_count += 1
                    continue

                img_data = requests.get(img_download_url, stream=True, timeout=15)
                if img_data.status_code == 200:
                    with open(file_path, 'wb') as f:
                        for chunk in img_data.iter_content(chunk_size=8192):
                            f.write(chunk)
                    downloaded_count += 1
                    print(f"  [{downloaded_count}] {filename}")

                time.sleep(0.5)

            except Exception as e:
                print(f"  Error on {link}: {e}")
                continue

        current_page += 1

    print(f"  Done: {downloaded_count} images → {output_dir}")


# ── Run ───────────────────────────────────────────────────────
for cat_name, cat_url in CATEGORIES.items():
    download_category_images(cat_name, cat_url)

print("\nAll downloads complete!")
for cat in CATEGORIES:
    path  = os.path.join(OUTPUT_BASE, cat)
    count = len(os.listdir(path)) if os.path.exists(path) else 0
    print(f"{cat}: {count} images")
```

```python
import os
import requests
from bs4 import BeautifulSoup
import time
import re

CATEGORIES = {
    "galaxies"    : "https://esahubble.org/images/archive/category/galaxies/",
    "nebulae"     : "https://esahubble.org/images/archive/category/nebulae/",
    "star_clusters": "https://esahubble.org/images/archive/category/starclusters/",
    "solar_system": "https://esahubble.org/images/archive/category/solarsystem/"
}

OUTPUT_BASE = "/content/hubble_dataset"
CDN_BASE    = "https://cdn.esahubble.org/archives/images/large"

def get_image_ids(soup):
    ids = []
    for a in soup.find_all('a', href=True):
        href = a['href']
        # must match /images/XXXX/ where XXXX is alphanumeric + hyphens, min 6 chars
        match = re.match(r'^/images/([a-zA-Z0-9\-]{6,})/$', href)
        if match:
            img_id = match.group(1)
            # explicitly exclude known navigation slugs
            exclude = ['viewall', 'search', 'potw', 'potm', 'archive', 'category']
            if img_id not in ids and img_id not in exclude:
                ids.append(img_id)
    return ids

def download_category_images(category_name, category_url):
    output_dir = os.path.join(OUTPUT_BASE, category_name)
    os.makedirs(output_dir, exist_ok=True)

    downloaded_count = 0
    current_page     = 1

    print(f"\n[+] Downloading: {category_name.upper()}")

    while True:
        page_url = f"{category_url}page/{current_page}/" if current_page > 1 else category_url
        response = requests.get(page_url, timeout=10)

        if response.status_code != 200:
            print(f"  Stopped at page {current_page} — status {response.status_code}")
            break

        soup     = BeautifulSoup(response.text, 'html.parser')
        img_ids  = get_image_ids(soup)

        if not img_ids:
            print(f"  No images on page {current_page}, done.")
            break

        print(f"  Page {current_page}: {len(img_ids)} images found")

        for img_id in img_ids:
            file_path = os.path.join(output_dir, f"{img_id}.jpg")

            if os.path.exists(file_path):
                downloaded_count += 1
                continue

            # directly construct CDN URL — no detail page needed
            cdn_url  = f"{CDN_BASE}/{img_id}.jpg"
            try:
                img_data = requests.get(cdn_url, stream=True, timeout=15)
                if img_data.status_code == 200:
                    with open(file_path, 'wb') as f:
                        for chunk in img_data.iter_content(chunk_size=8192):
                            f.write(chunk)
                    downloaded_count += 1
                    print(f"  [{downloaded_count}] {img_id}.jpg")
                else:
                    print(f"  CDN 404 for {img_id}")
                time.sleep(0.3)
            except Exception as e:
                print(f"  Error {img_id}: {e}")

        current_page += 1

    print(f"  Done: {downloaded_count} images → {output_dir}")


# ── Run ───────────────────────────────────────────────────────
for cat_name, cat_url in CATEGORIES.items():
    download_category_images(cat_name, cat_url)

print("\nAll downloads complete!")
for cat in CATEGORIES:
    path  = os.path.join(OUTPUT_BASE, cat)
    count = len(os.listdir(path)) if os.path.exists(path) else 0
    print(f"{cat}: {count} images")
```

```python
import os
import requests
import xml.etree.ElementTree as ET
import re
import time

CATEGORIES = {
    "galaxies"    : "https://esahubble.org/images/feed/category/galaxies/",
    "nebulae"     : "https://esahubble.org/images/feed/category/nebulae/",
    "star_clusters": "https://esahubble.org/images/feed/category/starclusters/",
    "solar_system": "https://esahubble.org/images/feed/category/solarsystem/"
}

OUTPUT_BASE = "/content/hubble_dataset"
CDN_BASE    = "https://cdn.esahubble.org/archives/images/large"

def get_all_image_ids(feed_url):
    ids = []
    resp = requests.get(feed_url, timeout=10)
    if resp.status_code != 200:
        print(f"Feed error: {resp.status_code}")
        return ids

    root = ET.fromstring(resp.content)
    for item in root.findall('.//item'):
        link = item.find('link')
        if link is not None and link.text:
            # extract ID from https://esahubble.org/images/potm2605a/
            match = re.search(r'/images/([a-zA-Z0-9\-]+)/$', link.text)
            if match:
                ids.append(match.group(1))

    return ids

def download_category(category_name, feed_url):
    output_dir = os.path.join(OUTPUT_BASE, category_name)
    os.makedirs(output_dir, exist_ok=True)

    print(f"\n[+] {category_name.upper()}")
    ids = get_all_image_ids(feed_url)
    print(f"  Found {len(ids)} image IDs in feed")

    downloaded = 0
    for img_id in ids:
        file_path = os.path.join(output_dir, f"{img_id}.jpg")
        if os.path.exists(file_path):
            downloaded += 1
            continue

        cdn_url = f"{CDN_BASE}/{img_id}.jpg"
        try:
            resp = requests.get(cdn_url, stream=True, timeout=15)
            if resp.status_code == 200:
                with open(file_path, 'wb') as f:
                    for chunk in resp.iter_content(chunk_size=8192):
                        f.write(chunk)
                downloaded += 1
                print(f"  [{downloaded}/{len(ids)}] {img_id}.jpg")
            else:
                print(f"  CDN 404: {img_id}")
            time.sleep(0.3)
        except Exception as e:
            print(f"  Error {img_id}: {e}")

    print(f"  Done: {downloaded} images → {output_dir}")

# ── Run ───────────────────────────────────────────────────────
for cat_name, cat_url in CATEGORIES.items():
    download_category(cat_name, cat_url)

print("\nAll downloads complete!")
for cat in CATEGORIES:
    path  = os.path.join(OUTPUT_BASE, cat)
    count = len(os.listdir(path)) if os.path.exists(path) else 0
    print(f"{cat}: {count} images")
```

```python
import os
import requests
import re
import time

# FIX 1: Removed hyphens. The archive slugs are 'starclusters' and 'solarsystem'
CATEGORIES = {
    "galaxies": "https://esahubble.org/images/archive/category/galaxies/",
    "nebulae": "https://esahubble.org/images/archive/category/nebulae/",
    "star_clusters": "https://esahubble.org/images/archive/category/starclusters/",
    "solar_system": "https://esahubble.org/images/archive/category/solarsystem/"
}

OUTPUT_BASE = "/content/hubble_dataset"
CDN_BASE    = "https://cdn.esahubble.org/archives/images/large"

def get_all_image_ids(category_url):
    ids = set()
    page = 1

    while True:
        # FIX 2: Correct Django routing path for pagination
        if page == 1:
            api_url = f"{category_url}?format=json"
        else:
            api_url = f"{category_url}page/{page}/?format=json"

        try:
            resp = requests.get(api_url, timeout=10)

            # 404 means we have successfully scraped the final page
            if resp.status_code == 404:
                break
            if resp.status_code != 200:
                print(f"  [-] Error {resp.status_code} on page {page}")
                break

            raw_text = resp.text
            matches = re.findall(r'/images/([a-zA-Z0-9\-]+)/', raw_text)

            page_ids = []
            for m in matches:
                # FIX 3: Valid IDs always contain numbers (e.g., heic1501a).
                # This filters out folders like 'potw', 'zip', 'archive' automatically!
                # We also explicitly ignore links starting with 'thumb'
                if any(c.isdigit() for c in m) and not m.lower().startswith('thumb'):
                    page_ids.append(m)

            if not page_ids:
                break

            initial_count = len(ids)
            ids.update(page_ids)

            if len(ids) == initial_count:
                break # Prevents infinite loops if the server stops advancing pages

            print(f"  -> Scanned page {page} | Total unique IDs harvested: {len(ids)}")
            page += 1
            time.sleep(0.2)

        except Exception as e:
            print(f"  [-] Parsing error on page {page}: {e}")
            break

    return list(ids)

def download_category(category_name, category_url):
    output_dir = os.path.join(OUTPUT_BASE, category_name)
    os.makedirs(output_dir, exist_ok=True)

    print(f"\n[+] {category_name.upper()}")
    ids = get_all_image_ids(category_url)
    print(f"  [!] Found {len(ids)} unique images. Starting download...")

    downloaded = 0
    for i, img_id in enumerate(ids, 1):
        file_path = os.path.join(output_dir, f"{img_id}.jpg")

        # Resume safety
        if os.path.exists(file_path):
            downloaded += 1
            continue

        cdn_url = f"{CDN_BASE}/{img_id}.jpg"
        try:
            resp = requests.get(cdn_url, stream=True, timeout=15)
            if resp.status_code == 200:
                with open(file_path, 'wb') as f:
                    for chunk in resp.iter_content(chunk_size=8192):
                        f.write(chunk)
                downloaded += 1
                print(f"  [{i}/{len(ids)}] Saved: {img_id}.jpg")
            else:
                # Silently skip CDN 404s (usually means the image is only available in .tif format)
                pass

            time.sleep(0.3)
        except Exception as e:
            print(f"  [-] Network error downloading {img_id}: {e}")

    print(f"  Done: {downloaded} images -> {output_dir}")

# ── Run ───────────────────────────────────────────────────────
if __name__ == "__main__":
    for cat_name, cat_url in CATEGORIES.items():
        download_category(cat_name, cat_url)

    print("\n[!] All downloads complete!")
    for cat in CATEGORIES:
        path  = os.path.join(OUTPUT_BASE, cat)
        count = len(os.listdir(path)) if os.path.exists(path) else 0
        print(f"{cat}: {count} total images stored")
```
